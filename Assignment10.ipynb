```python
#2.b. Create a histogram for each feature in the dataset.
corr = df.corr()
plt.subplots(figsize=(10,6))
sns.heatmap(corr, annot=True)
```

```python
#3.Create a box plot for each feature in the dataset
# importing packages

 
def graph(y):
 sns.boxplot(x="petal.width", y=y, data=df)
 
plt.figure(figsize=(10,10))
     
# Adding the subplot at the specified
# grid position
plt.subplot(221)
graph('sepal.length')
 
plt.subplot(222)
graph('sepal.length')
 
plt.subplot(223)
graph('sepal.length')
 
plt.subplot(224)
graph('sepal.length')
 
plt.show()
```

```python
# 4. Compare distributions and identify outliers.
# a.

sns.boxplot(x='sepal.width', data=df)
plt.show()
sns.boxplot(x='sepal.length', data=df)
plt.show()
sns.boxplot(x='petal.length', data=df)
plt.show()
sns.boxplot(x='petal.width', data=df)
plt.show()
```

#### In the above graph, the values above 4 and below 2 are acting as outliers in SepalWidthCm feature. Other features dont show outliers.
